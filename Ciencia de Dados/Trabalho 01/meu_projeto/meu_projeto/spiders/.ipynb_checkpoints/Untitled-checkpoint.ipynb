```python
import scrapy

class MunicipiosSpider(scrapy.Spider):
    name = "municipios"
    
    start_urls = urls
    
    def parse(self, response):
        prefeito = response.css(".prefeito dd::text").get()
        gentilico = response.css(".gentilico dd::text").get()
        area_territorial = response.css(".area-territorial dd::text").get()
        populacao_estimada = response.css(".estimativa-populacao dd::text").get()
        densidade_demografica = response.css(".densidade-populacional dd::text").get()
        escolarizacao = response.css(".indice-escolarizacao dd::text").get()
        idhm = response.css(".idhm-municipal dd::text").get()
        mortalidade_infantil = response.css(".mortalidade-infantil dd::text").get()
        receitas_realizadas = response.css(".receitas-realizadas dd::text").get()
        despesas_empenhadas = response.css(".despesas-empenhadas dd::text").get()
        pib_per_capita = response.css(".pib-per-capita dd::text").get()
        
        yield {
            "prefeito": prefeito,
            "gentilico": gentilico,
            "area_territorial": area_territorial,
            "populacao_estimada": populacao_estimada,
            "densidade_demografica": densidade_demografica,
            "escolarizacao": escolarizacao,
            "idhm": idhm,
            "mortalidade_infantil": mortalidade_infantil,
            "receitas_realizadas": receitas_realizadas,
            "despesas_empenhadas": despesas_empenhadas,
            "pib_per_capita": pib_per_capita,
            "url": response.url,
        }
        /html/body/main/section/div[2]/div/div[2]/div[2]/div[2]/ul/li[1]/div/p
```

```python
#from scrapy.crawler import CrawlerProcess
#from meu_projeto.spiders.municipios_spider import MunicipiosSpider

process = CrawlerProcess()
process.crawl(MunicipiosSpider)
process.start()
```
