```python
# Bibliotecas que serão utilizadas no trabalho

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sb

%matplotlib inline
```

```python
from sklearn.metrics import davies_bouldin_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Dataframe com os dados dos municípios cearenses

df = pd.read_csv('DF_Cidades.csv')
display(df)
```

```python
df.describe()
```

```python
# Calcular a mediana da variável "mortalidade_infantil"
mediana = df['mortalidade_infantil'].median()

# Substituir os valores faltantes pela mediana
df['mortalidade_infantil'].fillna(mediana, inplace=True)
```

```python
# Fazendo uma cópia do dataframe original

df_copia = df.copy()
df_copia
```

```python
# Excluíndo variáveis 

df_copia = df_copia.drop(columns = ['codigo','cidade', 'prefeito', 'gentilico', 'populacao_estimada', 
                                   'receitas_realizadas', 'despesas_empenhadas'])
```

```python
scaler = StandardScaler()
```

```python
df_copia[['area_territorial_T', 'densidade_demografica_T', 'escolarizacao_T', 'idhm_T', 'mortalidade_infantil_T', 
          'pib_T']] = scaler.fit_transform(df_copia[['area_territorial', 'densidade_demografica','escolarizacao', 
                                                     'idhm', 'mortalidade_infantil', 'pib']])
```

```python
df_copia
```

```python
# Create function to work out optimun number of clusters 

def optimise_k_means(data, max_k):
    means = []
    inertials = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters = k)
        kmeans.fit(data)
        
    # Generate the elbow plot
    fig = plt.subplots(figsize = (10, 5))
    plt.plot(means, inertials, 'o-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertials')
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(df_copia[['escolarizacao', 'pib']], 8)
```

```python
def get_kmeans_score(data, center):
    
    kmeans = KMeans(n_clusters=center, init='k-means++')
    model = kmeans.fit_predict(df_copia)
    
    # Calculate Davies Bouldin score
    score = davies_bouldin_score(df_copia, model)
    
    return score
```

```python
scores = []
centers = list(range(2,8))
for center in centers:
    scores.append(get_kmeans_score(df_copia, center))
    
plt.plot(centers, scores, linestyle='--', marker='o', color='b');
plt.xlabel('K');
plt.ylabel('Davies Bouldin score');
plt.title('Davies Bouldin score vs. K');
```

```python
# Dendogram for Heirarchical Clustering
import scipy.cluster.hierarchy as shc
from matplotlib import pyplot
pyplot.figure(figsize=(10, 7))  
pyplot.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(df_copia, method='ward'))
```

```python
def kmeans_(X, n_cluster):
    km = KMeans(
        n_clusters=n_cluster, init='k-means++'
    )

    y_km = km.fit_predict(X)
    
    for i in range(n_cluster):
        plt.scatter(
            X[y_km == i, 0], X[y_km == i, 1],
            #s=50, c='lightgreen',
            #marker='s', edgecolor='black',
            label='cluster '+str(i)
        )
    
    # plot the centroids
    plt.scatter(
        km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
        s=250, marker='*',
        c='red', edgecolor='black',
        label='centroids'
    )
    plt.legend(scatterpoints=1)
    plt.grid()
    plt.show()
```

```python
for i in range(2,6):
    kmeans_(X, i)
```
