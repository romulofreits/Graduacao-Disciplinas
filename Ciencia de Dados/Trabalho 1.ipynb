# Trabalho 1: Engenharia de Dados 

### Neste trabalho irei fazer uma raspagem de dados sobre Notebooks de 14 polegadas no site do Mercado Livre 

```python
# Importando a biblioteca de raspagem de dados
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link_dell = "https://lista.mercadolivre.com.br/informatica/portateis-acessorios/notebooks/dell/notebook-14-polegadas_NoIndex_True#applied_filter_id%3DBRAND%26applied_filter_name%3DMarca%26enhanced_position%3D0%26applied_value_id%3D8216%26applied_value_name%3DDell%26applied_value_order%3D2%26intervention_type%3DFILTER_INTERVENTION%26intervention_tracking_id%3D866f905c-2a73-4ac9-906f-29807216f971%26tracking_id%3D636213ee-99a5-49b4-bda7-56d88a8697c5"
```

```python
# Fazendo a requisição da página web por meio da biblioteca requests
requisicao = requests.get(link_dell) 
```

```python
print(requisicao) # se a requisição funcionar, ela retornará o número 200
# print(requisicao.text)
```

```python
# Parsear o HTML da page com a biblioteca BeautifulSoup
site_dell = BeautifulSoup(requisicao.text, 'html.parser') 
```

```python
# Agora a variável site_dell faz parte de uma classe do BeautifulSoup
type(site_dell)
```

```python
# Mostrando o HTML organizado
print(site_dell.prettify())
```

```python
# Vimos que na nossa estrutura html os nomes dos notebooks aprecem dentro da tag <h2>, logo vamos chamá-la

lista_notedell = site_dell.find_all("h2")
lista_notedell
```

```python
len(lista_notedell) # Temos 52 itens com a teg <h2>
```

```python
# Vimos que os nomes dos notebooks dell estão dentro da tag <h2> e dentro da classe "ui-search-item__title shops__item-title"

lista_notedell = site_dell.find_all("h2", {'class':"ui-search-item__title shops__item-title"})
lista_notedell
```

```python
len(lista_notedell) # Temos 52 itens com a teg <h2>
```

```python
# Pegando apenas o nome dos notebooks
lista_notedell[0].text
```

```python
# Vamos fazer um for para pegarmos o nome de todos os notebooks lenovo

for nome in lista_notedell:
    print(nome.text)
```

```python
valor_notedell = site_dell.find_all("div", {'class': 'ui-search-price ui-search-price--size-medium shops__price'})
valor_notedell
```

```python
len(valor_notedell)
```

```python
valor_notedell[0].text
```

```python
valor_notedell[0].text
```

```python
for preco in valor_notedell:
    print(preco.text)
```
